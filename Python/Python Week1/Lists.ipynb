```python
# 1
def List_Multiply(List:list) -> int:
    if not List:
        return 0

    product = 1
    for number in List:
        product *= number
    return product
```

```python
# 2
def List_Largest(List:list) -> int:
    if not List:
        return None

    # return max(List)

    largest = List[0]
    for number in List:
        if number > largest:
            largest = number
    return largest
```

```python
# 3
def List_Smallest(List:list) -> int:
    if not List:
        return None

    # return min(List)

    smallest = List[0]
    for number in List:
        if number < smallest:
            smallest = number
    return smallest
```

```python
# 4
def Tupple_Sort(List:list) -> list:
    if not List:
        return []

    sorted_list = sorted(List, key=lambda x: x[-1])
    return sorted_list
```

```python
# 5
def Remove_Duplicates(List:list) -> list:
    if not List:
        return []

    # return list(set(List))

    unique_list = []
    for item in List:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
```

```python
# 6
def Is_Empty(List:list) -> bool:
    return len(List) == 0
```

```python
# 7
def Count_Lowercase(List:list) -> list:
    if not List:
        return 0

    count = 0
    result = []
    for word in List:
        for char in word:
            if char.islower():
                count += 1
        result.append(count)
        count = 0
    return result
```

```python
# 8
def Extract_Consecutive(List:list, n:int) -> list:
    if not List or n <= 0:
        return []

    result = []
    count = 1

    for i in range(1, len(List)):
        if List[i] == List[i - 1]:
            count += 1
        else:
            if count == n:
                result.append(List[i - 1])
            count = 1

    if count == n:   # Checking the last one
        result.append(List[-1])

    return result
```

```python
# 9
def Largest_Odd(List:list) -> int:
    if not List:
        return "No odd numbers found"

    largest_odd = None

    for number in List:
        if number % 2 != 0 and (largest_odd is None or number > largest_odd):
                      largest_odd = number

    return largest_odd if largest_odd is not None else "No odd numbers found"
```

```python
# 10
def Remove_Specific_Elements(List:list) -> list:
    if not List:
        return []

    indices_to_remove = {0, 4, 5}
    result = []
    for i in range(len(List)):
        if i not in indices_to_remove:
            result.append(List[i])

    return result
```

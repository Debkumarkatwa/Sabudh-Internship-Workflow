```python
# 1
def Tuple_Create(*args) -> tuple:
    return tuple(args)
```

```python
# 2
def Print_Item(Tuple:tuple, index:int) -> any:
    if index < 0 or index >= len(Tuple):
        return None
    return Tuple[index]
```

```python
# 3
def Tuple_Add(Tuple:tuple, item:any) -> tuple:
    return Tuple + (item,)
```

```python
# 4
def Fourth_From_Last(Tuple:tuple) -> any:
    if len(Tuple) < 4:
        return None
    return Tuple[-4]
```

```python
# 5
def Tuple_To_Dict(Tuple:tuple) -> dict:
    if not Tuple:
        return {}
    result = dict()
    for i in Tuple:
        result[i[0]] = i[1]

    return result
```

```python
# 6
def Replace_Last_Value(Tuple_List:list, new_value:any) -> list:
    if not Tuple_List:
        return []

    result = []
    for t in Tuple_List:
        if len(t) < 1:
            result.append(t)
        else:
            new_tuple = t[:-1] + (new_value,)
            result.append(new_tuple)

    return result
```
